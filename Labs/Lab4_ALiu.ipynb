# Question 1

## Starbucks

```python
starbucks = requests.get("https://www.menuism.com/restaurant-locations/starbucks-coffee-39564")
soup = BeautifulSoup(starbucks.content, "html.parser")

SBrows = []

for location in soup.find_all("li", attrs={"class": ""})[17:68]:
    cells = location.find_all("a")
    
    name_tag = cells[0]
    nametext = name_tag.text
    namelist = nametext.split(" ")
    
    if len(namelist) == 5:
        name = namelist[0]
        number = namelist[4]
    elif len(namelist) == 6:
        name = namelist[0] + " " + namelist[1]
        number = namelist[5]
    else:
        name = namelist[0] + " " + namelist[1] + " " + namelist[2]
        number = namelist[6]
    number = number.replace("(", "") 
    number = number.replace(")", "")
    
    SBrows.append({
        "Name": name,
        "Number": number
    })
```

# Question 4

```python
wikipop = requests.get("https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population")
soup10 = BeautifulSoup(wikipop.content, "html.parser")

tables = soup10.find_all("table")
table = tables[0]
```

```python
poprows = []

for population in table.find_all("tr")[1:53]:
    
    cells = population.find_all("td", attrs = {"style" :"text-align: left;"})
    
    name_tag = cells[0].find("a") or cells[0]
    name = name_tag.text
    
    cell = population.find_all("td")
    
    population_tag = cell[3]
    
    population = population_tag.get_text(strip = True)
    
    poprows.append({
        "Name": name,
        "Population": population
    })

df2 = pd.DataFrame(poprows)

df2["Abbr"] = df2["Name"].apply(stateabb)
df2
```

```python
state_to_region = {
    'Northeast': ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont",
                  "New Jersey", "New York", "Pennsylvania"],
    'Midwest': ['Illinois', "Indiana", "Michigan", 'Ohio', "Wisconsin", "Iowa", "Kansas", "Minnesota",
               "Missouri", "Nebraska", "North Dakota", "South Dakota"],
    'South': ['Florida', 'Texas', "Delaware", "Georgia", "Maryland", "North Carolina", "South Carolina",
             "Virginia", "District of Columbia", "West Virginia", "Alabama", "Kentucky", "Mississippi",
             "Tennessee", "Arkansas", "Louisiana", "Oklahoma"],
    'West': ["Alaska", 'California', "Hawaii", "Oregon", "Washington", "Arizona",
            "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming"]
}

def map_state_to_region(state):
    for region, states in state_to_region.items():
        if state in states:
            return region
        
df3['Region'] = df3['State'].apply(map_state_to_region)
df3
```
